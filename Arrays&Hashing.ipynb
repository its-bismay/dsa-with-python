# 1. CONTAINS DUPLICATE (LC-217)

## my approach

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        myDict = {key: 0 for key in nums}

        for n in nums:
            if n in myDict:
                myDict[n] += 1
        
        if any (value > 1 for value in myDict.values()):
            return True
        else:
            return False
```

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        dup = set(nums)

        if len(dup) != len(nums):
            return True
        return False
```

# 2. VALID ANAGRAM (LC-242)

## my approach

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        myDict = {}
        for idx, char in enumerate(s):
            if char in myDict:
                myDict[char] += 1
            else:
                myDict[char] = 1
        
        for char in t:
            if char not in myDict:
                return False
                break
            else:
                if myDict[char] != t.count(char):
                    return False
                    break
        
        return True
```

## optimal approach (01):

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        myDict = {}
        for char in s:
            if char in myDict:
                myDict[char] += 1
            else:
                myDict[char] = 1
        
        for char in t:
            if char not in myDict or myDict[char] == 0:
                return False
            myDict[char] -= 1
        return True
```

## optimal approach (02):

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        counts = {}
        for i in range(len(s)):
            counts[s[i]] = counts.get(s[i], 0) + 1
            counts[t[i]] = counts.get(t[i], 0) - 1
        
        for val in counts.values():
            if val != 0:
                return False
        return True
```

# 3. TWO SUM (LC-1)

## my approach

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for idx1, num1 in enumerate(nums):
            for idx2, num2 in enumerate(nums):
                if idx1 <= idx2:
                    continue
                if num1+num2 == target:
                    return [idx1, idx2]
```

## optimal approach (01):

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        myDict = {}
        for idx, num in enumerate(nums):
            diff = target - num
            if diff in myDict:
                return [myDict[diff], idx]
            myDict[num] = idx
        return []
        
```

# 4. GROUP ANAGRAM (LC-49)

## my approach

```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        myDict = {}

        for n in strs:
            sortedString = "".join(sorted(n))
            if sortedString in myDict:
                myDict[sortedString].append(n)
            else:
                myDict[sortedString]=[n]

        finalList = []

        for key in myDict:
            finalList.append(myDict[key])

        return finalList
```

# 5. TOP K FREQUENT ELEMENTS (LC-347)

## my approach

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        myDict = {}
        for n in nums:
            if n not in myDict:
                myDict[n] = 1
            else:
                myDict[n] += 1

        myList=[[] for i in range (0,len(nums)+1)]

        for key, value in myDict.items():
            myList[value].append(key)
        
        ans=[]
        for i in range (len(myList)-1, 0, -1):
            for key, val in myDict.items():
                if val == i:
                    ans.append(key)
                    if len(ans) == k:
                        return ans
```
