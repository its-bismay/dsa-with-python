# CONTAINS DUPLICATE (LC-217)

## my approach

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        myDict = {key: 0 for key in nums}

        for n in nums:
            if n in myDict:
                myDict[n] += 1
        
        if any (value > 1 for value in myDict.values()):
            return True
        else:
            return False
```

## optimal aproach (01):

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:

        mySet = set()

        for n in nums:
            if n in mySet:
                return True
            mySet.add(n)
        
        return False
```

## optimal approach (02):

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        dup = set(nums)

        if len(dup) != len(nums):
            return True
        return False
```

# TWO SUM (LC-1)

## my approach

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for idx1, num1 in enumerate(nums):
            for idx2, num2 in enumerate(nums):
                if idx1 <= idx2:
                    continue
                if num1+num2 == target:
                    return [idx1, idx2]
```

## optimal approach (01):

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        myDict = {}
        for idx, num in enumerate(nums):
            diff = target - num
            if diff in myDict:
                return [myDict[diff], idx]
            myDict[num] = idx
        return []
        
```

# GROUP ANAGRAM (LC-49)

## my approach

```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        myDict = {}

        for n in strs:
            sortedString = "".join(sorted(n))
            if sortedString in myDict:
                myDict[sortedString].append(n)
            else:
                myDict[sortedString]=[n]

        finalList = []

        for key in myDict:
            finalList.append(myDict[key])

        return finalList
```

```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        myDict = {}

        for n in strs:
            sortedString = "".join(sorted(n))
            if sortedString in myDict:
                myDict[sortedString] = myDict.get(sortedString) + [n]
            else:
                myDict[sortedString]=[n]

        return list(myDict.values())
```
