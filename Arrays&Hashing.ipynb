# CONTAINS DUPLICATE (LC-217)

## my approach

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        myDict = {key: 0 for key in nums}

        for n in nums:
            if n in myDict:
                myDict[n] += 1
        
        if any (value > 1 for value in myDict.values()):
            return True
        else:
            return False
```

## optimal aproach (01):

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:

        mySet = set()

        for n in nums:
            if n in mySet:
                return True
            mySet.add(n)
        
        return False
```

## optimal approach (02):

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        dup = set(nums)

        if len(dup) != len(nums):
            return True
        return False
```
