## 2. LONGEST SUBSTRING WITHOUT REPEATING CHARACTERS (LC-3)

## my approach

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        uniqueChars = set()

        l = 0
        ans = 0

        for r in range(len(s)):
            while s[r] in uniqueChars:
                uniqueChars.remove(s[l])
                l += 1
            uniqueChars.add(s[r])
            ans = max(ans, r - l + 1)
```
